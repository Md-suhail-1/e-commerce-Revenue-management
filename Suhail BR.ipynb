### Ans 5

```python
# Adding new column revenue
def calcrevenue(disc_price):
    if(disc_price > 600):
        return (0.25 * disc_price)
    elif(disc_price > 350 and disc_price <= 600):
        return (0.15 * disc_price)
    elif(disc_price >100 and disc_price <= 350):
        return (0.10 * disc_price)
    else:b
        return (0.01 * disc_price)

ECom_Data['revenue'] = ECom_Data['discounted_price'].apply(calcrevenue)
```
